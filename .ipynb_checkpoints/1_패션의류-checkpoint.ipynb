```python
import requests
from copy import copy


def dfs(category_id):
    url = "https://api.itemscout.io/api/category/{0}/subcategories".format(category_id)
    response = requests.get(url = url)
    
    for d in response.json()['data']:
        d_id = d['id']
        d_is_leaf = d['is_leaf']
        
        if(d_is_leaf):
            leaf = copy(d_id)
            leaf_category.append(leaf)
            
        else:
            dfs(d_id)
    
paths = []
leaf_category = []
starting_category = [1,2,3,4,5,8,9] #출산/육아(6) , 식품 제외(7)

for i in starting_category:
    dfs(i)


print(leaf_category)

#1 : 패션의류
#2 : 패션잡화
#3 : 화장품/미용
#4 : 디지털/가전
#5 : 가구/인테리어
#6 : 출산/육아 (X)
#7 : 식품 (X)
#8 : 스포츠/레저
#9 : 생활/건강
```

```python
def get_keywords(data):
    res_keywords=[]
    
    for d in data:
        res_keywords.append(d['keyword'])
    
    return res_keywords

def remove_brands(keywords, brands):
    res_keywords = []
    
    for keyword in keywords:
        is_brand = 0
        
        for brand in brands:
            if keyword in brand:
                is_brand=1
                break
                
        if is_brand==0:
            res_keywords.append(keyword)
    
    return res_keywords


payload={
    'duration':'30d',
    'genders':'f,m',
    'age':'10,60'
}

for c in range(660,662):
    url_data = "https://api.itemscout.io/api/category/{0}/data".format(c)
    url_brands = "https://api.itemscout.io/api/category/{0}/brands".format(c)
    response_data = requests.post(url_data, json=payload)
    response_brands = requests.get(url_brands)
    data = response_data.json()['data']['data'].values()
    brands = response_brands.json()['data']
    
    print(brands)
    
    
    keywords = get_keywords(data)
    
    print(keywords.contains('딘트'))
    
    keywords_nb = remove_brands(keywords,brands)
    print(keywords_nb)
    print(len(keywords_nb))

    
    
```

```python
import requests
from copy import copy


def dfs(category_id, path):
    path.append(category_id)
    url = "https://api.itemscout.io/api/category/{0}/subcategories".format(category_id)
    response = requests.get(url = url)
    
    for d in response.json()['data']:
        d_id = d['id']
        d_is_leaf = d['is_leaf']
        
        if(d_is_leaf):
            path.append(d_id)
            p = copy(path)
            paths.append(p)
            path.pop()
            
        else:
            dfs(d_id,path)
    
    path.pop()

paths = []
path = []
starting_category = [1,2,3,4,5,8,9] #출산/육아(6) , 식품 제외(7)

for i in starting_category:
    dfs(i,path)


print(paths)

#1 : 패션의류
#2 : 패션잡화
#3 : 화장품/미용
#4 : 디지털/가전
#5 : 가구/인테리어
#6 : 출산/육아 (X)
#7 : 식품 (X)
#8 : 스포츠/레저
#9 : 생활/건강
```

```python
import os

f = open("path_category.txt",'w')
path_category = []

for p in paths:
    path_id = ''
    for n in p:
        path_id += str(n) + ','
    path_category.append(path_id[:-1])

f.writelines('\n'.join(path_category))
f.close()
```
